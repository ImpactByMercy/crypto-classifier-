```python
import pandas as pd

X = pd.read_csv("../data/processed/features.csv")
y_df = pd.read_csv("../data/processed/df_with_labels.csv")
```

```python
print(len(X), len(y_df))
# 1001, 1000
```

```python
X = X.iloc[:-1].reset_index(drop=True)
y_df = y_df.reset_index(drop=True)
```

```python
assert len(X) == len(y_df)
```

```python
mask = X.notna().all(axis=1)

X_clean = X[mask]
y_clean = y_df["label"][mask]
```

```python
print(X_clean.index[:5])
print(y_clean.index[:5])

print(X_clean.index[-5:])
print(y_clean.index[-5:])
```

```python
# Save features
X_clean.to_csv(
    "../data/processed/X_clean.csv",
    index=False
)

# Save labels
y_clean.to_csv(
    "../data/processed/y_clean.csv",
    index=False,
    header=True  # important: keep column name
)

print("Saved X_clean.csv and y_clean.csv")
print("X shape:", X_clean.shape)
print("y shape:", y_clean.shape)
```
